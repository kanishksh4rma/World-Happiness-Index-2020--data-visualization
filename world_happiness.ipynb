# data visualization

```
sns.pairplot(data = df)
```

```
fig,ax = plt.subplots(figsize=(12,8))
with sns.axes_style('white'):
  sns.kdeplot(data = df.upperwhisker,data2=df.lowerwhisker,shade=True,ax=ax)
plt.show()
```

```
with sns.axes_style('white'):
  sns.jointplot(x='Generosity',y='Ladder score',data = df, kind='reg' , color='#FF5733')
```

```
sns.scatterplot(x='Explained by: Log GDP per capita',y='Explained by: Social support',data = df,color='#2f8df1')
```

```
fig,ax = plt.subplots(figsize=(8,6))
sns.scatterplot(x=df['Freedom to make life choices'] , y=df['Social support'],color='#D73E3E')
plt.show()
```

```
fig,ax = plt.subplots(figsize=(15,45))
sns.barplot(y=df.index,x=(df['Healthy life expectancy']),orient='h')
plt.show()
```

```
fig,ax = plt.subplots(figsize=(10,7))
sns.distplot(df['Logged GDP per capita'],color='#E0811F')
plt.axvline(df['Logged GDP per capita'].mean())
plt.show()
```

```
fig,ax = plt.subplots(figsize=(10,7))
sns.distplot(df['Ladder score'],color='#3466CF')
plt.axvline(df['Ladder score'].mean())
plt.show()
```

```
fig,ax = plt.subplots(figsize=(10,7))
sns.violinplot(df.drop(['Regional indicator'],axis=1))
plt.show()
```

```
sns.lmplot(x='Logged GDP per capita',y='Freedom to make life choices',data = df, height=4,col='Regional indicator',ci=None,col_wrap=4,line_kws={'color':'red'},scatter_kws={ 'color' : 'blue' })
```

```

sns.lmplot(x='Dystopia + residual',y='Ladder score in Dystopia',data = df, height=4,col='Regional indicator',ci=None,col_wrap=4,line_kws={'color':'red'},scatter_kws={ 'color' : 'green' })
```
